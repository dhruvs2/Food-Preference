```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Food_Preference.csv")
```

```python
df = df.drop(['Timestamp','Participant_ID'], axis = 1) # Not Needed for prediction
```

```python
list(df.Nationality).count('Indian') # Roughly 83% Indians!
```

```python
Des = df.Dessert.unique() 
Foo = df.Food.unique()
Jui = df.Juice.unique() 
print(Des,Foo,Jui)
```

# Plotting Session

```python
fig, ax = plt.subplots(1,3, figsize = (12,5))
sns.countplot(df['Gender'], ax=ax[0])
sns.countplot(df['Juice'], ax=ax[1])
sns.countplot(df['Food'], ax=ax[2])
fig.tight_layout()
plt.figure(figsize = (2,5))
plt.show()
```

```python
sns.catplot(x="Food", y="Age", hue="Nationality", kind="bar", data=df,aspect = 2.5)
```

```python
fig = plt.figure(figsize=(10,20))
ax1 = fig.add_subplot(1,1,1)
sns.countplot(y=df['Age'],ax=ax1)
```

```python
plt.figure(figsize = (15,6))
sns.stripplot(x="Food", y="Age", data=df, jitter = 0.1, hue = "Gender") # Jitter distinguishes overlapping points
```

```python
plt.figure(figsize = (15,6))
sns.boxplot(x="Age", y="Food", data=df,hue = "Gender")
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x="Age", y="Juice", data=df,hue = "Gender")
```

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1,1,1)
sns.boxplot(x=df['Age'],y=df['Dessert'])
```
